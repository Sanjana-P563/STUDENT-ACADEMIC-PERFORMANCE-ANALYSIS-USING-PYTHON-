3.Motivation Level vs Exam Score-Shows how student motivation impacts performance.Higher motivation levels often cluster at higher exam scores, which gives a strong positive correlation.

4.Teacher Quality vs Exam Score-Shows the effect of teacher quality on student scores.Higher quality categories with higher scores suggest that better teaching leads to better student outcomes.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a cross-tabulation of Peer_Influence and Motivation_Level
contingency_table = pd.crosstab(data['Peer_Influence'], data['Motivation_Level'])

# Create a stacked bar chart
contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Chart of Peer Influence vs. Motivation Level')
plt.xlabel('Peer Influence')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Motivation Level')
plt.show()
```

Interpretation:

The stacked bar chart shows the relationship between Peer Influence and Motivation Level.If â€œHigh Peer Influenceâ€ bars have a larger portion of high motivation, it suggests peer influence may positively affect motivation.
Conversely, a large portion of low motivation in â€œLow Peer Influenceâ€ may indicate less motivated students tend to have weaker peer influence.

The overall bar height shows how many students fall into each peer influence category, which is useful for understanding the dataset composition.
